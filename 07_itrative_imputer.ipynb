2. Multiple imputation by chained Equations(MICE)

```python
import pandas as pd
import numpy as np
import seaborn as sns
# itrative imputer is to impute all the null values together rather then individually
from sklearn.experimental import enable_iterative_imputer
# itrative imputer is just an experimental, it might be give the wrong value because it is in beta version means under improving 
from  sklearn.impute import IterativeImputer
# first point is enable the itrative imputer first then import the IterativeImputer without enable the ItrativeImputer will not import or not work
df = sns.load_dataset('titanic')
df.head()

# missing_values = df.isnull().sum().sort_values(ascending=False)
# print(missing_values)
```

```python
# check the missing,null,nan values from the data
df.isnull().sum().sort_values(ascending=False)
```

Enode the data into  0,1,2

```python
# LabelEncoder is used to convert the catagrical values into numeric in the form of 0 1 2 0 1 2 0 1 2 
from sklearn.preprocessing import LabelEncoder
# mention the columns the to encode 
columns_to_encode = ['sex','embarked','who','deck','class','embark_town','alive']
# Dictionary to store labelEncoders for each column
label_encoders = {}
# Loop to apply LabelEncoder to impute each column 
for col in columns_to_encode:
    # call the LabelEncoder and store in an object
    le = LabelEncoder()
    # train the model and tranform
    df[col] = le.fit_transform(df[col])
    # store the encoder into the label_encoders dictionary
    label_encoders[col] = le
df.head()
```

impute the missing values by using the IterativeImputer

```python
# impute the missing values by using the IterativeImputer
# call the IterativeImputer class with max_iteratiion = 10
# means it will run 10 time for checking and then calculate the average and then impute on the missing_values
imputer = IterativeImputer(max_iter=10)

# impute the missing values using the IterativeImputer in a for loop for age, embark_town, embarked and deck columns
# columns to impute 
column_to_impute = ['age','embarked','embark_town','deck']

# Loop to impute each column
for col in column_to_impute:
    df[col] = imputer.fit_transform(df[[col]])

# ceck the missing missing_values
df.isnull().sum().sort_values(ascending = False)
```

Inverse the Encoded data by using the inverse_transform into same as orignal form

```python
# inverse the transform for encoded columns 
for col in columns_to_encode:
    # Retrieve the corresponding LabelEncoder for  the column 
    le = label_encoders[col]
    # Inverse transform the data and convert to integer type
    df[col] = le.inverse_transform(df[col].astype(int))

df.head()
```
