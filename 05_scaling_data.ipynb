import Libraries and scalar  techniques

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder,StandardScaler, MinMaxScaler
```

read the csv file

```python
df = pd.read_csv('All CSV files/employee.csv')
print(df.columns)
```

```python
df.head()
```

call and fit+transform stander scalar model

```python
stds = StandardScaler()
df['std_scaler'] = stds.fit_transform(df[['JoiningYear']])
df.head()
```

call and fit_transform MinMax scalar model

```python
min_max_scaler = MinMaxScaler()
df['min_max_scaler'] = min_max_scaler.fit_transform(df[['JoiningYear']])
```

```python
df[['min_max_scaler','std_scaler']].head()
```

Import the MaxAbsScalar 

```python
from sklearn.preprocessing import MaxAbsScaler
```

call and fit_transform MaxAbs scalar model

```python
max_abs_scaler = MaxAbsScaler()
df['max_abs_scaler'] = max_abs_scaler.fit_transform(df[['JoiningYear']])
df[['max_abs_scaler']].head()
```

```python
df[['std_scaler','min_max_scaler','max_abs_scaler']].head()
```
