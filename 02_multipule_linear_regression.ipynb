Import libraries, Encoder, Preprocessing and Model

```python
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
```

Load dataset

```python
# load the data
df = sns.load_dataset('tips')
df.head()
```

```python

# # make the X and y 
# X = df[['size','sex','day','time']]
# y = df['total_bill']
```

Call the Linear Model and Encode the data from categorical to numeric form

```python
# call the models
model = LinearRegression()

# encode the sex,day,time
columns = ['sex','time','day']
encoder_columns = {}
for col in columns:
    encode = LabelEncoder()
    df[col] = encode.fit_transform(df[col])
    encoder_columns[col] = encode
```

Make Features and Labels

```python
# make the X and y 
X = df[['size','sex','day','time']]
y = df['total_bill']
```

Fit the model

```python
predict = model.fit(X,y)
df.head()
```

Check the Coefficient and intercept

```python
# now check the coffients in multiple linear regression
print("Coefficients : ",predict.coef_)

# now check the intercept of the multiple linear Regression
print("Intercept : ",predict.intercept_)
```

Predict the totall bill

```python

# now predict the total bill 

print("Prediction of the total bill after : ",predict.predict([[4,1,2,0]]))
```

Inverse the encoded data 

```python
# inverse the encoded columns
for col in columns:
    le = encoder_columns[col]
    df[col] = le.inverse_transform(df[col].astype(int))
```
