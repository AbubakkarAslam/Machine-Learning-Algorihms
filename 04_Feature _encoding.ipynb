Import libraries Label Encoder for preprocessing

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
```

read the employee csv file

```python
df=pd.read_csv("All CSV files/employee.csv")
df.head()
```

Label Encoding by using the Technique of "LABEL ENCODER"

```python
le = LabelEncoder()
df['eb']=le.fit_transform(df['EverBenched'])
df.head()
```

Feature Encoding by using the Technique of "ORDER" (feature having ordinal Categories)

```python
education_encoder = {'Bachelors':0,'Masters':1,'phD':2}
df['edu_encoder'] = df["Education"].map(education_encoder)
df['edu_encoder'].head()
```

ONE-HOT Encoding

```python
df = pd.get_dummies(df,columns = ['City'],prefix = "City")
df.head()
```

Frequency Encoding

```python
education_counts = df['Education'].value_counts().to_dict()
df['frequency_encoding'] = df["Education"].map(education_counts)
df.head()
```

Target Encoding

```python
target_encoding = df.groupby('Gender')["eb"].mean().to_dict()
df["targrt_encoding"] = df["Education"].map(target_encoding)
df
```
