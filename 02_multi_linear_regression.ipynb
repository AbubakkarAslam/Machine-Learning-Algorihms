Import libraries, Encoder, Preprocessing and Model

```python

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
```

Load dataset

```python
# load the data
df = pd.read_csv("House Price Prediction Dataset.csv")
df.head()
```

```python
df.isnull().sum().sort_values(ascending = True)
```

Call the Linear Model and Encode the data from categorical to numeric form

```python
# call the models
model = LinearRegression()

columns_to_encod = ['Location','Condition','Garage']
encoded_columns= {}
for col in columns_to_encod:
    label = LabelEncoder()
    df[col] = label.fit_transform(df[col])
    encoded_columns[col] = label
df.head()
```

Make Features and Labels

```python
# make the X and y 
X = df.drop(['Price','Area','Id'],axis=1)
y = df.Price
```

Fit the model

```python
predict = model.fit(X,y)
df.head()
```

Check the Coefficient and intercept

```python
# now check the coffients in multiple linear regression
print("Coefficients : ",predict.coef_)

# now check the intercept of the multiple linear Regression
print("Intercept : ",predict.intercept_)
```

```python
X.columns
```

Predict the totall price of house

```python

# now predict the total price of house 

import numpy as np
new_data = np.array([[5, 4, 2, 2020, 1, 0, 1]])
pred = model.predict(new_data)
# or
# pred = model.predict([[5, 4, 2, 2020, 1, 0, 1]])
print("Prediction of the total price of the house is : ",pred)
```

Inverse the encoded data 

```python
# inverse the encoded columns

for col in columns_to_encod:
    en = encoded_columns[col]
    df[col] = en.inverse_transform(df[col])

```
