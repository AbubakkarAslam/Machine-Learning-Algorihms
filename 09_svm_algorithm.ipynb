```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
dir(iris)
```

```python
iris.feature_names
```

Make Dataframe by using the pandas

```python
df = pd.DataFrame(iris.data, columns=(iris.feature_names))
df.head()
```

Add iris target column into the dataframe

```python
df['target']=iris.target
df.head()
```

check the iris target names which is there in  0,1,2

```python
iris.target_names
```

check the target values in 0

```python
df[df.target==0].head()
```

check the target values in 1

```python
df[df.target==1].head()
```

check the target values in 2

```python
df[df.target==2].head()
```

make column with name target_flower and assign the 0,1,2 with names

```python
df['target_flower'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

import the matplotlib for scatter plot

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

make dataframe from df and assign the targeted columns to the new dataframes for scatter plots

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
df.head()
```

 make scatter plot with the dataframe df0 

```python
plt.xlabel('sepal lenght (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'], color='green',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color= 'red', marker= '.')
```

make scatter plot with dataframe with name df1

```python
plt.xlabel('petal lenght (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'], color='orange',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color= 'blue', marker= '.')
```

```python
df.head()
```

call the model SVC and add the labels and features also split the data in 80% and 20%

```python
model = SVC(kernel = 'linear',gamma = 100)
X = df.drop(['target','target_flower'], axis = 1)
y = df['target']
X_test, X_train, y_test, y_train = train_test_split(X,y,test_size = 0.2)

print("lenth",len(X_train))
print("lenth",len(X_test))
```

fit the model

```python
model.fit(X_train,y_train)
```

check the score of trained model

```python
model.score(X_test,y_test)
```

save the csv file 

```python
df.to_csv("../All CSV files/iris.csv", index=False)
```
