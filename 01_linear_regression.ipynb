Import the libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import seaborn as sns
```

Import Linear Regression model form scikit learn

```python
from sklearn.linear_model import LinearRegression
```

Load dataset from seaborn libraries

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

Assign feature and label to "X" and "y"

```python
X  = df[['total_bill']]
y = df['tip']
```

make scatter plot for view the total bill and tips in graph

```python
sns.scatterplot(x='total_bill',y='tip', data=df)
```

Plot the Linear Model plot

```python
sns.lmplot(x='total_bill',y='tip',data=df)
```

Call the Linear Regression model 

```python
model = LinearRegression()
```

Fit the model

```python
model.fit(X,y)
```

Pedict the tips

```python

model.predict([[50]])
```
